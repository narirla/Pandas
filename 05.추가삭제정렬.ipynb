```python
import pandas as pd
import numpy as np
```

```python
data =  {'eng':[10,30,50,70,60],
         'kor':[20,40,60,80,75],
         'math':[90,50,20,70,30]}
df = pd.DataFrame(data, index=['a','b','c','d','e'] )
df
```

### 추가

### 컬럼 추가

```python
df['test'] = [1,2,3,4,5]  # 컬럼이 有면 수정, 無면 추가
```

```python
df['eng'] + df['kor']
```

```python
df['sum'] = df['eng'] + df['kor']
```

```python
df.insert(1, 'test1',[1,2,3,4,5]) # df.insert(추가할 컬럼인덱스(위치), 추가할 인덱스 명, 추가할 값)
```

### 행추가

```python
df.loc['f'] = [1,2,3] # 해당 인덱스가 有면 수정, 無면 추가
```

### 삭제
- drop( index=[부여된인덱스명,...], colums=[부여된컬럼명,....] )

```python
# df = df.drop( index=['a','c','d'])
# df.drop( index=['a','c','d'], inplace=True) # inplace=True 를 주게 되면 삭제된 값이 바로 반영됨
df.drop( index=['a','c','d'] )
```

```python
df.drop( columns=['eng','math'])
```

```python
df.drop(index=['a','d'], columns=['kor']) # drop은 행열 모두 삭제 가능
```

```python
df.drop(index=['a','d'], axis=0) #axis=0(행) a와 d를 인덱스명으로 인식하여 삭제함
```

```python
df.drop(columns=['kor'], axis=1) #axis=1 컬럼으로 인식하여 'kor'을 삭제
```

### 정렬

```python
df.sort_values(by='math') #기준 컬럼을 必 지정해줘야함
```

```python
df.sort_values(by='math', ascending=False) #내림차순 
```

```python
df.loc['c','kor'] = 40
```

```python
df.sort_values(by=['kor','math']) # 1차 정렬은 국어로 하고 2차 정렬은 수학으로 정렬
```

```python
df.sort_index(ascending=False) #내림차순
```

```python
df.sort_index(axis=1, ascending=False) #인덱스를 컬럼 정렬해서 알파벳 역순으로 정렬
```
