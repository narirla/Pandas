```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x=['홍길동','이순신','임꺽정','이이']
y=[90,77,50,80]
plt.bar( x, y)
plt.show()
```

```python
x=['홍길동','이순신','임꺽정','이이']
y=[90,77,50,80]
plt.barh( x, y, color=['r','g','b','y'])
plt.show()
```

```python
import pandas as pd
```

```python
xname=['홍길동','이순신','임꺽정','이이']
x = pd.Series( [1,2,3,4] )
kor = [ 90,77,50,80 ]
eng = [60,20,50,20]
plt.xticks( x+0.2, xname)
plt.bar( x,kor,width=0.4) #bar가 겹치지 않도록 0.4 추가
plt.bar( x+0.4,eng,width=0.4)
plt.show()
```

```python
### stacked bar : daframeplot( stacked=True)
data =  {'eng':[10,30,50,70],
         'kor':[20,40,60,80],
         'math':[90,50,20,70]}
df = pd.DataFrame(data, index=['a','b','c','d'] )
df.plot( kind='bar',stacked=True)
```

```python
xname=['홍길동','이순신','임꺽정','이이']
x = pd.Series( [1,2,3,4] )
kor = [ 90,77,50,80 ]
eng = [60,20,50,20]
plt.xticks( x, xname)
plt.bar( x,kor)
plt.bar( x,eng, bottom=kor)
plt.legend(['국어','영어'])
# plt.text( 1, 90/2, '90점',va='center',ha='center' )
# plt.text( 1, 90+60/2, '60점',va='center',ha='center' )
# plt.text( 1, 60, '60점' )

for a, k,e in zip(x, kor, eng):
    plt.text(a, k/2, f'{k}점', va='center', ha='center')
    plt.text(a, k+e/2, f'{e}점', va='center', ha='center')

plt.show()
```
