```python
xname=['','','','','']
x =[1,2,3,4,5]
y1 =[10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xticks(x, xname, rotation=45)
plt.xlabel('x')
plt.ylabel('y')
plt.title('')
plt.grid(True)
plt.plot( x, y1,'ro--', x, y2,'bs-', x,y3, 'g<-.')
plt.legend(['','','']) #   
plt.show()
```

```python
xname=['','','','','']
x =[1,2,3,4,5]
y1 =[10,14,19,20,25]
plt.xticks(x, xname, rotation=45)
plt.plot( x, y1, 'ro--')
# plt.text(1,10,'10' )
# plt.text(2,14,'14' )
plt.ylim(0,100)
for a,b in zip(x,y1):
    plt.text(a,b,f'{b}' )
plt.show()
```

```python
x =[1,2,3,4,5]
y1 =[10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig, axes = plt.subplots(2,2, figsize=(7,7))
# axes
axes[0,0].plot( x,y1)
axes[0,1].plot( x,y1)
axes[1,0].plot( x,y1)
plt.show()
```

```python
x =[1,2,3,4,5]
y1 =[10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig = plt.figure( figsize=(7,7))
ax1 = fig.add_subplot( 221 ) # (2,2,1)
ax2 = fig.add_subplot( 222 )
ax3 = fig.add_subplot( 223 )
ax1.plot( x,y1)
ax2.plot( x,y2)
ax3.plot( x,y3)
plt.show()
```

```python
x =[1,2,3,4,5]
y1 =[10,14,19,20,25]
y2 = [5,8,10,25,20]
fig, axes = plt.subplots()
axes.set_xlabel( 'x')
axes.set_ylabel( 'y',color='b')
axes.plot( x,y1)

ax2 = axes.twinx()
ax2.set_ylabel( 'y1',color='r')
ax2.plot( x,y2,'r--')
plt.show()
```
