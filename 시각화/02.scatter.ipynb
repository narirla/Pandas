```python
height = [155,160,165,172,177,180,185]
weight = [43,55,62,64, 67,74,90]

표준체중(kg)=(신장(cm)-100)×0.85
비만도(%)=현재체중/표준체중×100
비만도가
90이하:저체중, (y,100)
90∼110:정상,  (b ,150)
110∼120:과체중, (c, 200)
120이상:비만    (r ,300)
    
그래프각점에 비만도결과값(저체중, 정상,과체중, 비만)글자를
표시하고 원의 크기는  각 결과 값의 색상과 크기로
그리시요.
```

```python
for w,h in zip(weight,height):
    standardW = (h-100) * 0.85;
    heavy = w/standardW * 100
    if heavy >=120:
        plt.scatter(h,w,s=300,c='r')
        plt.text(h,w,'비만')
    elif (heavy >= 110) & (heavy < 120):
        plt.scatter(h,w,s=200,c='c')
        plt.text(h,w,'과체중')
    elif (heavy >= 90) & (heavy < 110):
        plt.scatter(h,w,s=150,c='b')
        plt.text(h,w,'정상')
    else:
        plt.scatter(h,w,s=100,c='y')
        plt.text(h,w,'저체중')
plt.show()
```

```python
result =[]
color =[]
size=[]
def obesityFn( h, w):
    standardW = (h-100) * 0.85;
    heavy = w/standardW * 100
    if heavy >=120:
        result.append( '비만')
        color.append('r')
        size.append(300)
    elif 110<= heavy< 120:
        result.append( '과체중')
        color.append('c')
        size.append(200)
    elif 90<= heavy< 110:
        result.append( '정상')
        color.append('b')
        size.append(150)
    else:
        result.append( '저체중')
        color.append('y')
        size.append(100)        
```

```python
height = [155,160,165,172,177,180,185]
weight = [43,55,62,64, 67,74,90]
for w,h in zip(weight,height):
    obesityFn(h,w)

plt.scatter(height,weight,s=size,c=color)
for x,y,r in zip( height, weight, result):
    plt.text( x,y,r)
plt.show()
```
