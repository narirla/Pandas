```python
import pandas as pd
import numpy as np
```

```python
data = {'aa':10, 'bb':20, 'cc':30, 'dd':40, 'ee': 50}
sr = pd.Series(data, name= '국어점수')
sr
```

### 수정

```python
sr[0] = 100 
# sr['aa'] = 100
# sr.aa = 100
# sr.iloc[0] = 100 # 권장
# sr.loc['aa'] = 100 # 권장
```

```python
sr[1:4] = [1,2,3]
# sr[ 'bb':'dd' ] = [1,2,3]
# sr.iloc[1:4] = [1,2,3]
# sr.loc['bb':'dd'] = [1,2,3] 
```

```python
sr
```

```python
### Q) 30이상인 데이터에 5를 더한 값으로 수정하시오.
sr[sr>=30] = sr[sr>=30]+5 
```

### 추가

```python
sr.loc['ff'] = 66 # 해당 인덱스가 있으면 수정, 없으면 추가
```

```python
sr
```

### 삭제
sr.drop( [부여된 인덱스명] )
- numpy 
np.delete(arr,[인덱스,..])

```python
# sr.drop(['aa','dd']) #적용
sr.drop(['aa','dd'], inplace=True) #바로 삭제 적용
```

```python
### Q) 25 이상인 데이터를 삭제하시오
sr.drop(sr[sr>=25].index) # 부여된 인덱스명을 가져와야 함으로 .index로 부여된 인덱스명 추출
```

```python
### Q) 25 이상인 데이터의 총합
sr[sr>=25].sum()
```
