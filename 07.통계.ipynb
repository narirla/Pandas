### 조건별
- apply

```python
def fn (v):
    print( 'v=>', v ) #v= 컬럼별 시리즈
    print('='*30)
    return v.sum()
```

```python
df.apply(fn) # 각 컬럼별 총합
```

```python
df.apply(fn, axis=1) # 행별 총합
```

```python
df.apply(lambda v: v.sum())
```

```python
df.apply(lambda v: v.sum(), axis=1)
```

```python
df.apply(lambda v: '합격' if v.mean()>=70 else '불합격', axis=1)
```

```python
# 국어,영어, 수학 점수의 평균이 70점 이상이면 
# '합격' 아니면 '불합격' 으로 결과를 출력하시오
df['결과'] = df.apply(lambda v: '합격' if v.mean()>=70 else '불합격', axis=1)
```
