# 누적합, 누적평균

##  1. `cumsum()` : 누적합 (지금까지 값을 계속 더함)

### 원리
- 첫 번째 값부터 현재 행까지 **모든 value를 더한 값**을 기록
- 즉, `sum(value[:i])`
value = [10, 20, 30, 40]

| value | cumsum |
|------|--------|
| 10   | 10     |
| 20   | 30     | (10+20)
| 30   | 60     | (10+20+30)
| 40   | 100    | (10+20+30+40)

---
##  2. `expanding().mean()` : 누적평균 (지금까지의 평균)

###  원리  
- 첫 번째 값부터 현재 행까지 **모든 value의 평균**을 기록  
- 즉, `mean(value[:i])`

value = [10, 20, 30, 40]

| value | expanding mean |
|-------|----------------|
| 10    | 10.0           | 평균 = 10
| 20    | 15.0           | 평균 = (10+20)/2
| 30    | 20.0           | 평균 = (10+20+30)/3
| 40    | 25.0           | 평균 = (10+20+30+40)/4

#  누적합(cumsum)·누적평균(expanding mean)

데이터 분석에서 두 함수는 “시간 흐름에서 변화를 이해하는 핵심 도구”

---
#  1. 누적합(cumsum)
### **시간이 지날수록 얼마나 쌓였는가?** 를 보기 위해 사용


### ① 전체 추세 파악  
- 매출이 계속 증가하는 구조인가?
- 방문자 수가 매일 쌓여서 큰 성장을 하고 있는가?
- 캠페인 효과가 시간과 함께 누적되고 있는가?

### ② 목표 달성 추적(Progress Tracking)
- 예: 월 매출 목표 1억  
  → `cumsum`을 보면 “얼마나 달성했는지” 바로 확인 가능

### ③ 이벤트의 누적 영향력 분석  
- 광고, 프로모션 후 매일 조금씩 증가한 효과가  
  → 시간이 지날수록 얼마나 누적되었는지 파악 가능
---

# ex) 매출 누적합  
value = 일 매출  
→ `cumsum`을 보면 **"이번 달 총매출이 어떻게 만들어져 가는지"** 분석 가능.

---

#  2. 누적평균(expanding().mean)
###  **데이터가 안정되는 방향(장기 평균)**을 확인하기 위해 사용  
특히 처음 숫자가 요동칠 때 “전체적인 기준선”을 찾는 데 매우 유용.


### ① 장기 트렌드(평균 기준선) 파악  
- 시험 점수 → 계속 변하지만 누적평균은 “실력의 기준점”을 보여줌  
- 일 매출 → 들쭉날쭉하지만 누적평균은 “안정적인 평균값”을 향해 수렴  

### ② 변동성(Volatility) 감소 확인  
- 데이터가 시간이 지나면서 안정되는지?  
- 아니면 계속 불안정한지?

### ③ 시스템의 정상상태(Steady-state) 확인  
예: 공장 생산량  
- 초반에는 불안정  
- 누적평균이 일정해지면 “시스템이 안정됨”

---


| 분석 목적 | 추천 함수 |
|--------|-------------|
| 총 매출/총 방문자 등 **누적 성과 분석** | `cumsum` |
| 초반 노이즈 제거 후 **장기 평균·기준선 파악** | `expanding().mean` |
| 추세 변화 감지(단기) | rolling().mean |
| 변동성·이상치 영향 제거 | ewm().mean |

---

- **누적합(cumsum)** → *“지금까지 얼마나 쌓였는가?”*  
- **누적평균(expanding mean)** → *“전체적으로 어느 수준에 머무를 것인가?”*
