```python
def fn(v):
    print(v)
    print('='*20)
    return v.mean()
```

```python
df.groupby('학년')['중간'].agg(fn) # 1학년 2학년 따로 출력
```

```python
df.groupby('학년')['중간'].agg( lambda v:v.mean() )
```

```python
# Q) 학년 별로 중간, 기말고사 평균이 70점 이상이면 합격 아니면 불합격
df.groupby('학년')[['중간','기말']].agg( lambda v: '합격' if v.mean()>=70 else '불합격' )
```

```python
df.pivot_table(values='중간', index='학년') #학년별 중간고사 집계
```

```python
df.pivot_table(values='중간', index='학년', aggfunc="sum")
```

```python
df.pivot_table(values=['중간','기말'] , 
               index='학년', aggfunc=["mean","sum"], sort=False) #학년별 중간 기말에 대한 평균, 총합의 값
```

```python
df.pivot_table(values='중간', index='학년',
                aggfunc=lambda v: '합격' if v.mean() >= 70 else '불합격')
```

```python
df1 = df.pivot_table(values='중간', index='학년',
                aggfunc=[ 'mean',lambda v: '합격' if v.mean() >= 70 else '불합격'])
df1.rename(columns={'mean':'평균', '<lambda>':'결과'} )
```

```python
df2 = df.groupby('학년')['중간'].agg( ['mean', lambda v: '합격' if v.mean()>=70 else '불합격'] )
df2.rename(index={'1학년':'1학년임'}, 
            columns={'mean':'중간고사 평균','<lambda_0>':'중간고사 결과'})
```
