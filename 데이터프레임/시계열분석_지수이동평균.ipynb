### 지수 이동 평균(EMA)
---
```text
가까운(최근) 데이터에 더 큰 비중을 주고,
오래된 데이터는 점점 적게 반영해서
전체 흐름(추세)을 부드럽게 보여주는 방법입.

1. EMA의 필요성

시계열 데이터(매출, 주가, 트래픽 등)는 대부분
위로 갔다 내려갔다 요동치는 그래프임
이건 너무 들쭉날쭉해서 추세를 보기가 어렵다.
흐름을 부드럽게 만드는 ‘평균선’을 만드는 기법이 필요.

2. 일반 이동평균(MA) vs 지수 이동평균(EMA)
MA(이동평균): 단순 평균

최근 데이터건 오래된 데이터건 똑같은 비중
→ 모든 날의 중요도가 같음

EMA(지수 이동평균)

최근 데이터일수록 더 중요한 정보라고 생각해서
가중치(weight)를 더 많이 줌
오래된 데이터는 자동으로 비중이 작아짐

즉, 최근 변화를 더 민감하게 반영함

3. 예시

가정:
날짜별 매출: [100, 102, 130, 90, 95]

✔ 이동평균(MA)은 그냥 평균
(100 + 102 + 130) / 3 = 110.6

✔ EMA는 이렇게 생각함
3일 전 데이터: 중요도 10%
2일 전: 30%
어제 데이터: 60%

“어제 데이터가 제일 중요해!” 라고 말하는 셈
그래서 EMA는 실제 값에 더 가까운,
빠르게 반응하는 부드러운 선을 만들어줌.

4. EMA의 장점?

빠른 추세 변화 감지
갑자기 오르기 시작하는 시점
갑자기 떨어지기 시작하는 시점
→ MA보다 빨리 잡아냄
노이즈(잡음) 제거: 데이터를 부드럽게 다듬어줌

주가·트래픽·매출 등 거의 모든 시계열 분석에서 기본적으로 사용

5. 파이썬으로 EMA 구하는 방법
df['EMA_14'] = df['value'].ewm(span=14).mean()
span=14 = 최근 14일 중심으로 가중치 부여
ewm() = 지수 가중 이동 평균 함수

6. 요약
EMA = “최근 데이터를 더 중요하게 보는 부드러운 평균선”
추세 판단,변화 감지,노이즈 제거에 아주 강력하다.
