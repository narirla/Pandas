```python
import pandas as pd
import numpy as np
```

```python
data =  {'eng':[10,30,50,70,60],
         'kor':[20,40,60,80,75],
         'math':[90,50,20,70,30]}
df = pd.DataFrame(data, index=['a','b','c','d','e'] )
df
```

```python
df+2
```

```python
df+df
```

```python
df + [10,20,30] #브로드 캐스팅으로 확장되어 행열로 더 해짐
```

```python
df['eng'] = df['eng'] + 2 #수정:요소별로 할당되어 수정 됨
```

```python
df.loc['a'] = df.loc['a'] + 2 #a행이 기존 값이 2를 더한 값으로 수정됨
```

```python
df.loc['c':'d','kor':] = df.loc['c':'d','kor':] + 2
```

## boolean indexing
- 행단위: [[True,False,...]]

```python
df[[True,False,False,True,True]]
```

```python
df['eng'] >30
```

```python
df[df['eng'] >30] #df[[False,False,True,True,True]]
```

```python
df[ (df['eng'] == 30) | (df['eng'] == 70) ] # 영어 점수가 30 또는 70인 행
```

```python
df[df['eng'].isin([30,70])] # 위 값과 동일
```

```python
df['eng']
```

```python
df[(df['eng'] >= 30) & (df['eng'] <= 70)]
```

```python
df[ df['eng'].between(30,70) ]
```
