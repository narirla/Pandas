### 퀴즈
1. 3개의 데이터프레임을 하나로 병합(concat)한 후 아래의 문제를 해결하시오.
2. 년도별 사고(건) 평균을 구하시오.
3. 2016년과 2015년 사고를 대비할 때 가장 많이 증가한 월을 구하시오. 2016-2015 max인 것의 월 도출


```python
# encoding: utf-8, euc-kr,cp949(행이 많은 경우)
sleep2014 =pd.read_csv('data/2014년 졸음운전 교통사고.csv', encoding='euc-kr')
sleep2015 =pd.read_csv('data/2015년 졸음운전 교통사고.csv', encoding='euc-kr')
sleep2016 =pd.read_csv('data/2016년 졸음운전 교통사고.csv', encoding='euc-kr')
```

```python
# 1. 3개의 데이터프레임을 하나로 병합(concat)한 후 아래의 문제를 해결하시오.
sleepDF = pd.concat( [sleep2014,sleep2015,sleep2016]) 
sleepDF
```

```python
sleepDF['년도'] = sleepDF['구분'].str[:5]
sleepDF['월'] = sleepDF['구분'].str[5:]
```

```python
# 2. 년도별 사고(건) 평균을 구하시오.
sleepDF.groupby('년도')[['사고(건)']].mean()
```

```python
s2015 = sleepDF[sleepDF['년도']=='2015년'][['월','사고(건)']]
s2016 = sleepDF[sleepDF['년도']=='2016년'][['월','사고(건)']]
s2015.index = s2015['월']
s2016.index = s2016['월']
(s2016['사고(건)'] - s2015['사고(건)']).nlargest(1, keep='all').index
# sr = (s2016 - s2015)
# sr
```

```python
s2015 = sleepDF[sleepDF['년도']=='2015년'][['월','사고(건)']]
s2016 = sleepDF[sleepDF['년도']=='2016년'][['월','사고(건)']]
s2015.set_index('월', inplace=True) #특정 값을 인덱스로 지정
s2016.set_index('월', inplace=True)
# s2016.reset_index() 0,1,2,3....으로 되돌림
```
