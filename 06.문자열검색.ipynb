```python
import pandas as pd
import numpy as np
```

```python
sr = pd.Series(['홍길동','이순신','김철수','김순이','이홍김'])
sr.index=['aa','bb','cc','dd','ee']
sr
```

### 데이터가 문자열 : str
- sr.str
- sr.index.str

```python
print(type(sr.str))
```

```python
sr.str[0]
```

```python
sr.str[-1]
```

```python
sr.str[1:]
```

```python
sr.index.str[0]
```

```python
sr[sr.str.contains('홍')] #where name rlike '홍' 
```

```python
sr[sr.str.contains('^홍')] # 홍으로 시작
```

```python
sr[sr.str.contains('김$')] # 김으로 끝나는 것
```

```python
sr[sr.str.contains('[홍이]')] # 홍 또는 이
```

```python
sr[sr.str.contains('[홍이]순')] #contains을 통해 정식 패턴 매칭
```

```python
sr[sr.str.contains('길동|순이|이홍')] # 또는
```

```python
sr[sr.index.str.contains('a')] # 인덱스에 a라는 글자가 있는 것
```

### 문자열 바꾸기

```python
sr.replace('홍길동',' 테스트') # sr.replace는 정확하게 일치해야 변경 가능, 정규식 패턴 미적용
```

```python
# sr.str.replace('홍','테스트') # 정규식 패턴 적용
sr.str.replace('김$','테스트',regex=True) # 정규식 패턴 적용 -> 김으로 끝나는것 만 변경
```
