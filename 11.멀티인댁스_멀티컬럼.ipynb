```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family']='NanumGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
np.random.seed(100)
data = np.random.randint( 30 , 101, size=(9,6))
```

```python
df = pd.DataFrame(data)
df
```

```python
df['년도'] = [2023,2023,2023,2024,2024,2024,2025,2025,2025]
df['학년'] = [ '1학년','2학년','3학년','1학년','2학년','3학년','1학년','2학년','3학년']
```

```python
# 멀티 인덱스 : 두개 이상의 컬럼을 인덱스로 선택 
df.set_index( ['년도','학년'] , inplace=True)
```

```python
df.reset_index(inplace=True) # 인덱스를 다시 복원
```

```python
df.index = pd.MultiIndex.from_product([[2023,2024,2025],['1학년','2학년','3학년']])
```

```python
df.drop(columns=['년도','학년'],inplace=True)
```
