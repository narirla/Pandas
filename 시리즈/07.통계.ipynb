### 조건별 apply

```python
def fn(v):
    # print('v =', v)
    # print('='*10)
    # if v<=30:
    #     return v+1
    # else:
    #     return v+2
    return v+1 if v<=30 else v+2
```

```python
sr.apply(fn)
```

```python
sr.apply(lambda v: v+1 if v<=30 else v+2)
```

```python
### Q) 40점 이상이면 합격 아니면 불합격 인 시리즈를 구하시오.
sr.apply(lambda v: '합격' if v>=40 else '불합격')
```
