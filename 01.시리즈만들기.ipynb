                    시계열 데이터, 다양한 데이터 읽을 수 有, 시각화)

```python
import pandas as pd
import numpy as np
```

- [10,20,30]
- (10,20,30)
- { 'aa' :10, 'bb' :20, 'cc' : 30 }
- np.array( [10,20,30] ) 
- 리스트, 튜플 , 딕셔너리 다 1차원

```python
d1 = [10,20,30]
d2 = (10,20,30)
d3 = { 'aa':10, 'bb':20, 'cc':30 } 
```

```python
sr = pd.Series(d1 , index=['aa', 'bb', 'cc'],
                dtype=np.int32, name='국어 점수') #인덱스에 이름을 줄 수 있음, 라벨링 가능
sr
```

```python
sr1 = pd.Series(d2)
sr1
```

```python
sr2 = pd.Series(d3)
sr2
```

```python
sr3 = pd.Series(np.array([11,22,33]))
sr3
```
