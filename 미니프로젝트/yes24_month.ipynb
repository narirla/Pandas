```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By


def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    #     chrome_options.add_argument('headless')
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), options=chrome_options
    )
    return driver


driver = set_chrome_driver()
```

```python
url = "https://www.yes24.com/product/category/monthbestseller?categoryNumber=001&pageNumber=1&pageSize=50&type=month&saleYear=2025&saleMonth=12"
driver.get(url)
```

```python
bestsellers = []

url = "https://www.yes24.com/product/category/monthbestseller?categoryNumber=001&pageNumber=1&pageSize=50&type=month&saleYear=2025&saleMonth"
year = 2025

for month in range(1, 13):
    
    month = month
    site_url = f"{url}={month}"
    
    print(f"\n--- {month}월 데이터 수집 중 ---")
    
    try:
        driver.get(site_url)
        
        for i in range(1, 51): 
            try:
                ranking = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[1]/div[1]/div/em').text.strip()
                title = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[2]/a[1]').text.strip()
                author = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[3]/span[1]').text.strip().replace(' 저', '').replace(' 글', '').replace(' 원', '').replace(' 역', '').replace(' 정보 더 보기', '').replace(' 편', '').split('/')[0].strip().replace('저','').replace('그림','').replace('공','')
                publisher = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[3]/span[2]/a').text.strip()
                publish_date = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[3]/span[3]').text.strip().replace(' ','')
                price = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[4]/strong/em').text.strip().replace(',','')
                sale = driver.find_element(By.XPATH,f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[4]/span[1]/em').text.strip()
                try:
                    review_count = driver.find_element(By.XPATH, f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[5]/span[2]/a/em[2]').text.strip().replace(',', '')
                except:
                    try:
                        review_count = driver.find_element(By.XPATH, f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[5]/span[1]/a/em[2]').text.strip().replace(',', '')
                    except:
                        review_count = 0
                try:
                    score = driver.find_element(By.XPATH, f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[5]/span[3]/em').text.strip()
                except:
                    try:
                        score = driver.find_element(By.XPATH, f'//*[@id="yesBestList"]/li[{i}]/div/div[2]/div[5]/span[2]/em').text.strip()
                    except:
                        score = 0
                data = {
                    'year': int(year),
                    'month': int(month),
                    'rank' : int(ranking),
                    'title' : title,
                    'author' : author,
                    'publisher' : publisher,
                    'publish_date' : publish_date,
                    'price' : int(price),
                    'discount' : int(sale),
                    'review_cnt' : int(review_count),
                    'score' : float(score),
                    'site' : '예스24'
                }
                
                bestsellers.append(data)
                
            except Exception as e:
                print(e)

    except Exception as e:
        print(e)
        
print(f"\n{len(bestsellers)}개 데이터")
```
